```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
data = np.random.rand(10, 2)

# Perform hierarchical clustering
linked = linkage(data, 'single')

# Plot dendrogram
dendrogram(linked, labels=np.arange(1, 11))
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 8, 18, 32, 50])

# Create an interpolation function (linear interpolation in this case)
linear_interp = interp1d(x, y, kind='linear')

# Generate new points for interpolation
x_interp = np.linspace(1, 5, 100)
y_interp = linear_interp(x_interp)

# Plot the original data and the interpolated curve
plt.scatter(x, y, label='Original Data')
plt.plot(x_interp, y_interp, label='Linear Interpolation')
plt.title('Linear Interpolation Example')
plt.legend()
plt.show()
```

```python
from scipy.io import savemat, loadmat

# Save data to a MATLAB file
data_to_save = {'x': x, 'y': y}
savemat('data.mat', data_to_save)

# Load data from the MATLAB file
loaded_data = loadmat('data.mat')
loaded_x = loaded_data['x']
loaded_y = loaded_data['y']

print("Loaded x:", loaded_x)
print("Loaded y:", loaded_y)
```
