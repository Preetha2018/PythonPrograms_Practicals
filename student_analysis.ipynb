```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data into a Pandas DataFrame
df = pd.read_csv("studentmarks.csv")

# Display the first few rows of the DataFrame
print(df.head())

# Get information about the DataFrame
print(df.info())

# Summary statistics
print(df.describe())

```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values if necessary
# Example: df = df.dropna()
```

```python
# Average marks of each subject
subject_avg = df.mean()

# Total marks of each student
df['Total Marks'] = df.sum(axis=1)

# Highest and Lowest marks
highest_marks = df.max()
lowest_marks = df.min()

# Top performing students
top_students = df.nlargest(5, 'Total Marks')

# Distribution of marks using a histogram
plt.hist(df['Total Marks'], bins=20, edgecolor='black')
plt.title('Distribution of Total Marks')
plt.xlabel('Total Marks')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Bar chart for average marks of each subject
subject_avg.plot(kind='bar', color='skyblue')
plt.title('Average Marks of Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.show()

# Pie chart for the percentage of students in different grade ranges
grade_bins = [0, 40, 60, 75, 100]
grade_labels = ['Fail', 'Pass', 'Credit', 'Distinction']
df['Grade'] = pd.cut(df['Total Marks'], bins=grade_bins, labels=grade_labels)
grade_counts = df['Grade'].value_counts()
grade_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['red', 'orange', 'yellow', 'green'])
plt.title('Distribution of Grades')
plt.show()
```
