```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

weather_data =  pd.read_csv('Weather_Data.csv')

# Displaying the first few rows of the DataFrame
print(weather_data.head())

```

```python
# Get basic information about the dataset
print(weather_data.info())

# Check for missing values
print(weather_data.isnull().sum())

# Descriptive statistics
print(weather_data.describe())
```

```python
#findind uniuq values and count of unique values
weather_data['Weather'].unique()
weather_data.nunique()
weather_data.count()
```

```python
#groupby
weather_data.groupby('Weather').get_group('Snow').shape
```

```python
# Find the number of times when the wind speed was exactly 4 km/hr?
print("wind speed is exactly 4km/hr")
weather_data[weather_data['Wind Speed_km/h'] == 4].shape
```

```python
#What is the mean Visibility of a given dataset
print("mean visiblity")
weather_data['Visibility_km'].mean()
```

```python
# Analyze temperature trends
average_temperature = weather_data['Temp_C'].mean()
temperature_range = weather_data['Temp_C'].max() - weather_data['Temp_C'].min()

# Analyze precipitation
total_precipitation = weather_data['Dew Point Temp_C'].sum()
average_precipitation = weather_data['Dew Point Temp_C'].mean()

print("average temperature ", average_temperature)
print("temperature range ", temperature_range)
print("Total preception",total_precipitation)
print("average preception ", average_precipitation)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Load the provided weather data
weather_data = {
    'Date/Time': ['01-01-2012 00:00', '01-01-2012 01:00', '01-01-2012 02:00', '01-01-2012 03:00'],
    'Temp_C': [-1.8, -1.8, -1.8, -1.5],
    'Dew Point Temp_C': [-3.9, -3.7, -3.4, -3.2],
    'Rel Hum_%': [86, 87, 89, 88],
    'Wind Speed_km/h': [4, 4, 7, 6],
    'Visibility_km': [8, 8, 4, 4],
    'Press_kPa': [101.24, 101.24, 101.26, 101.27],
    'Weather': ['Fog', 'Fog', 'Freezing Drizzle,Fog', 'Freezing Drizzle,Fog']
}
# Assuming 'Date/Time' is in datetime format
weather_data['Date/Time'] = pd.to_datetime(weather_data['Date/Time'], format='%d-%m-%Y %H:%M')

# Line plot for temperature trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date/Time', y='Temp_C', data=weather_data, marker='o', label='Temperature (C)')
plt.title('Temperature Trends Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()

# Line plot for precipitation (Dew Point Temp) trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date/Time', y='Dew Point Temp_C', data=weather_data, marker='o', color='orange', label='Dew Point Temperature (C)')
plt.title('Dew Point Temperature Trends Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Dew Point Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Load the provided weather data
weather_data = {
    'Date/Time': ['01-01-2012 00:00', '01-01-2012 01:00', '01-01-2012 02:00', '01-01-2012 03:00'],
    'Temp_C': [-1.8, -1.8, -1.8, -1.5],
    'Dew Point Temp_C': [-3.9, -3.7, -3.4, -3.2],
    'Rel Hum_%': [86, 87, 89, 88],
    'Wind Speed_km/h': [4, 4, 7, 6],
    'Visibility_km': [8, 8, 4, 4],
    'Press_kPa': [101.24, 101.24, 101.26, 101.27],
    'Weather': ['Fog', 'Fog', 'Freezing Drizzle,Fog', 'Freezing Drizzle,Fog']
}

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Date/Time' is in datetime format
weather_data['Date/Time'] = pd.to_datetime(weather_data['Date/Time'], format='%d-%m-%Y %H:%M')

# Histogram for temperature distribution
plt.figure(figsize=(10, 6))
sns.histplot(weather_data['Temp_C'], bins=20, kde=True, color='skyblue')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()

# Bar chart for weather types
plt.figure(figsize=(10, 6))
sns.countplot(y='Weather', data=weather_data, order=weather_data['Weather'].value_counts().index, palette='viridis')
plt.title('Distribution of Weather Types')
plt.xlabel('Count')
plt.ylabel('Weather Type')
plt.show()

# Scatter plot for Wind Speed vs Visibility
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Wind Speed_km/h', y='Visibility_km', data=weather_data, s=100, hue='Visibility_km', palette='muted')
plt.title('Wind Speed vs Visibility')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Visibility (km)')
plt.legend(title='Visibility (km)')
plt.show()
```
