```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]

# Plotting the data
plt.plot(x_values, y_values, label='Sample Line')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sample Line Graph')

# Adding legend
plt.legend()

# Display the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 22, 35, 28],
    'Salary': [50000, 60000, 45000, 70000, 55000],
    'Department': ['HR', 'IT', 'Marketing', 'Finance', 'IT']
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Displaying the dataset
print("Original Dataset:")
print(df)
print("\n")

# Basic statistics
print("Basic Statistics:")
print(df.describe())
print("\n")

# Plotting a bar chart
df.plot(kind='bar', x='Name', y='Salary', legend=False)
plt.title('Salary Distribution')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
file_path = 'sales_data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("Original Dataset:")
print(df.head())
print("\n")

# Basic statistics
print("Basic Statistics:")
print(df.describe())
print("\n")

# Total revenue per product
total_revenue_per_product = df.groupby('Product')['Revenue'].sum().reset_index()

# Plotting a bar chart for total revenue per product
total_revenue_per_product.plot(kind='bar', x='Product', y='Revenue', legend=False)
plt.title('Total Revenue per Product')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.show()
```
